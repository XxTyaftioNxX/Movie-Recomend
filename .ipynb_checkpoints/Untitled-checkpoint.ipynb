```python
import pandas as pd
import numpy as np
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = 'API_KEY'
from tmdbv3api import Movie

tmdb_movie = Movie() 
def get_titles(x):
    result = tmdb_movie.search(x)
    if not result:
      return np.NaN
    else:
      movie_id = result[0].id
      response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
      data_json = response.json()
      if data_json['original_title']:
          return data_json['original_title']
      else:
          return np.NaN  
```

```python
movies = pd.read_csv('main_data.csv')
```

```python
movies.drop('summary', axis=1, inplace=True)
```
