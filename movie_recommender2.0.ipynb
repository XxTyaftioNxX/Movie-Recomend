```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('main_data.csv')
```

```python
movies
```

```python
#dropping summary
movies.drop('summary', inplace=True, axis=1)
```

```python
#removing spaces for more accurate results
movies['actor_1_name'] = movies['actor_1_name'].str.replace(' ', '')
movies['actor_2_name'] = movies['actor_2_name'].str.replace(' ', '')
movies['actor_3_name'] = movies['actor_3_name'].str.replace(' ', '')
movies['director_name'] = movies['director_name'].str.replace(' ', '')
```

```python
#adding tag
movies['tags'] = movies['overview'] + ' ' + movies['genres'] + ' ' + movies['actor_1_name'] + ' ' + \
movies['actor_2_name'] + ' ' + movies['actor_3_name'] + ' ' + movies['director_name']
```

```python
movies['tags'] = movies['tags'].str.lower()
```

```python
movies
```

```python
#gathering only required data
req_data = movies[['movie_title', 'tags']]
```

```python
req_data
```

```python
req_data.isnull().sum()
```

```python
req_data.dropna(inplace=True)
```

```python
req_data.isna().sum()
```

### stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
            
    return " ".join(y)
```

```python
#applying stemming to the tags
req_data['tags'] = req_data['tags'].apply(stem)
```

### Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
count_matrix = cv.fit_transform(req_data['tags'])
```

```python
count_matrix.toarray()
```

```python
cv.get_feature_names()
```

### cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(count_matrix)
```

```python
def recommend(name):
    user_movie_index = req_data[req_data['movie_title'] == name].index[0]
    print(user_movie_index)
    recommended_movies = list(enumerate(similarity[user_movie_index]))
    sorted_recommended_movies = sorted(recommended_movies, key=lambda x: x[1], reverse=True)[1:20]
    
    for i in sorted_recommended_movies:
        print(req_data.iloc[i[0]].movie_title)
```

```python
recommend('Spectre')
```
